Logistic Map Simulation

```python
import matplotlib.pyplot as plt
import scipy as sp

def logistic_map(r, x0, n):
    xs = [x0]
    for i in range (n):
        x0 = x0*r*(1-x0)
        xs.append(x0)
    return xs

r = 4
x0 = 0.5
n = 100

xs = logistic_map(r, x0, 100)
plt.plot(xs)    
```

```python
from scipy.fft import fft, fftfreq
import numpy as np
tim = np.arange(0, 101, 1)
N = len(xs)
y = fft(xs)[:N//2]
f = fftfreq(N, np.diff(tim)[0]) [:N//2]
plt.plot(f, np.abs(y))
```

Logistic Map Bifurcation diagram

```python
R = np.linspace(0.1, 4, 10000)

Y = []
for r in R:
    x = np.random.random()
    for n in range (100):
        x = r*x*(1-x)
    for n in range(100):
        x = r*x*(1-x)
    Y.append(x)
plt.plot(R, Y, ls = "", marker = ",")
plt.show()            
    
```

Lyapunov‑exponent estimation (simple tangent‑map method)

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map(x, r):
    return r*x*(1-x)
def d_logistic_map(x, r):
    return r*x*(1-2*x)

r = 4.0
x = 0.5
delta = 1e-8
n_steps = 1000

lyapunov_sum = 0.0

for i in range (n_steps):
    x = logistic_map(x, r)
    df = d_logistic_map(x, r)
    delta = df * delta
    lyapunov_sum += np.log(abs(df)) #growth factor at each step
    
lyap_exp = lyapunov_sum / n_steps
print("Lyapunov Exponent:", lyap_exp)
```
